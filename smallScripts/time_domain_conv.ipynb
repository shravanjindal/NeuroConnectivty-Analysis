```python
import numpy as np
import matplotlib.pyplot as plt
def generate_morlet_wavelet(frequency, sampling_rate, duration, w0=5):
    t = np.arange(-duration / 2, duration / 2, 1 / sampling_rate)
    sine_wave = np.exp(1j * 2 * np.pi * frequency * t)
    gaussian_window = np.exp(-0.5 * (t * w0) ** 2)
    morlet_wavelet = sine_wave * gaussian_window
    return morlet_wavelet

def time_domain_convolution(signal, kernel):
    return np.convolve(signal, kernel, mode='full')

# Example usage:
# Define the signal
signal = np.random.randn(100)  # Example signal

# Generate a sine wavelet kernel
frequency = 5  # Frequency of the sine wavelet in Hz
sampling_rate = 100  # Sampling rate in samples per second
duration = 1  # Duration of the wavelet in seconds
sine_wavelet = generate_morlet_wavelet(frequency, sampling_rate, duration)

# Perform convolution
convolved_signal = time_domain_convolution(signal, sine_wavelet)

# plotting
plt.plot(signal)
plt.show()
plt.plot(sine_wavelet)
plt.show()
plt.plot(convolved_signal)
plt.show()
```
